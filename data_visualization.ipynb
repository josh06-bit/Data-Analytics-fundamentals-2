### Types of Data Visualization in Matplotlib

### 1. Line Plot

The line plot, a cornerstone of data visualization, excels at revealing trends over time or across different categories. It connects data points with a line, visually depicting how a value changes in relation to another variable.

```python
# Data for number of orders per month
data = {
    "Months": ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"],
    "No. Of Orders for Jacket": [8, 10, 12, 14, 18, 20, 26, 24, 22, 14, 12, 9],
    "No. Of Orders for Summer Clothes": [25, 27, 24, 20, 15, 13, 9, 13, 14, 16, 18, 22]
}

# Create a DataFrame
order_per_month_df = pd.DataFrame(data)
order_per_month_df
```

```python
# markers: o, s, ^, <, >, x, +
# linestyle: solid, dashed, dotted, dashdot, none
plt.plot(order_per_month_df["Months"], order_per_month_df["No. Of Orders for Jacket"], 
         marker="o" ,color="blue", linestyle="dashed", linewidth=2, label="No. Of Orders for Jacket")
plt.plot(order_per_month_df["Months"], order_per_month_df["No. Of Orders for Summer Clothes"], 
         marker="s" , color="red", linestyle="dotted", linewidth=3, label="No. Of Orders for Summer Clothes")

# Adding a label
plt.xlabel("Months", fontsize=12, fontweight="bold")
plt.ylabel("No. Of Orders", fontsize=12, fontweight="bold")

# Adding title
plt.title("No. of Orders per Month", fontsize=14, fontweight="bold")

# Add the legend
plt.legend(loc="center right", fontsize=8, bbox_to_anchor = (1.5, -0.1))
# upper right, upper left, lower right, lower left
# center right, center left
# bbox_to_anchor makes the legends move freely by x,y coordinates


# Add annotations near the markers
"""for i, month in enumerate(order_per_month_df["Months"]):
    plt.annotate(f"{order_per_month_df['No. Of Orders for Jacket'][i]}", (order_per_month_df["Months"][i], order_per_month_df["No. Of Orders for Jacket"][i]),
    fontsize=8, color="blue")
    plt.annotate(f"{order_per_month_df['No. Of Orders for Summer Clothes'][i]}", (order_per_month_df["Months"][i], order_per_month_df["No. Of Orders for Summer Clothes"][i]), fontsize=8, color="orange")
"""
#grid lines
plt.grid(True)

# Display the lineplot
plt.show()
```

### 2. Pie Charts

Pie charts, those delightful circles sliced into wedges, are a go-to choice for visualizing proportions of a whole.  Matplotlib's pie chart functionality, accessible through the pyplot.pie() function, empowers you to represent data categories and their relative sizes in a clear and intuitive way.

```python
# Data for spending amounts for each category
data = {
    "Categories": ['Rent', 'Food', 'Transportation', 'Savings', 'Other'],
    "Amounts": [500, 300, 200, 100, 100]
}

spending_amount_df = pd.DataFrame(data)
spending_amount_df
```

```python
# Create a Pie Chart
plt.pie(spending_amount_df["Amounts"], labels=spending_amount_df["Categories"], 
autopct="%.2f%%", colors=["#8ecae6", "#219ebc", "#023047", "#ffb703", "#fb8500"],
explode=[0.05,0.05,0.05,0.05,0.05])
# Hexadecimal: 0-9, A-F
# #RRGGBB

# Add a Title
plt.title("Budget for each Category")

# Add a legend
plt.legend(labels=spending_amount_df["Categories"], title="Categories", bbox_to_anchor=(1.5, 0.5))

plt.show()
```

#### 3. Bar Charts

Bar charts, the workhorses of data visualization, excel at showcasing comparisons between different categories.  Matplotlib's bar chart functionality, primarily accessed through the pyplot.bar() function, allows you to translate numerical data into a series of rectangular bars. The height of each bar corresponds to the value it represents, while the bars are positioned along a horizontal axis that typically represents the categories being compared.

```python
# Data for student enrollment per course
data = {
    "Course": ['Data Analytics', 'Frontend Development', 'DevOps', 'Full-Stack Development'],
    "Students": [30, 25, 32, 28]
}

students_per_course_df = pd.DataFrame(data)
students_per_course_df
```

```python
# Control the width and height
plt.figure(figsize=(8, 5))

# Create a Bar Chart
plt.bar(students_per_course_df["Course"],
        students_per_course_df["Students"],
        color= ["#5fad56", "#f2c14e", "#f78154", "#4d9078"],
        width=0.8
        )
# width: width of the bar [0 - 1]
plt.title("Students enrolled per Course", fontsize=14, fontweight="bold")
plt.xlabel("Courses", fontsize=12, fontweight="bold")
plt.ylabel("No. of Students", fontsize=12, fontweight="bold")
plt.xticks(rotation=45)
plt.show()
```
